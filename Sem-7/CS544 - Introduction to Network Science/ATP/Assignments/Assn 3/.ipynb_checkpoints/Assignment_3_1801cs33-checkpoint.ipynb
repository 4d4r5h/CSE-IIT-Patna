```python
# Importing necessary libraries
import re
import pandas as pd
import bs4
import requests
import spacy
from spacy import displacy
nlp = spacy.load('en_core_web_sm')

from spacy.matcher import Matcher 
from spacy.tokens import Span 

import networkx as nx

import matplotlib.pyplot as plt
from tqdm import tqdm

pd.set_option('display.max_colwidth', 200)
%matplotlib inline
```

```python
# import wikipedia sentences
candidate_sentences = pd.read_csv("wiki_sentences_v2.csv")
candidate_sentences.shape
```

```python
candidate_sentences.loc[0]['sentence']
```

```python
# For extracting the entities
def get_entities(sent):
  # Finding named entities in sent
  doc = nlp(sent)
  entity_names = [ent.text for ent in doc.ents]
  entity_labels = [ent.label_ for ent in doc.ents]
  ## chunk 1
  ent1 = ""
  ent2 = ""

  prv_tok_dep = ""    # dependency tag of previous token in the sentence
  prv_tok_text = ""   # previous token in the sentence

  prefix = ""
  modifier = ""

  #############################################################
  
  for tok in nlp(sent):
    ## chunk 2
    # if token is a punctuation mark then move on to the next token
    if tok.dep_ != "punct":
      # check: token is a compound word or not
      if tok.dep_ == "compound":
        prefix = tok.text
        # if the previous word was also a 'compound' then add the current word to it
        if prv_tok_dep == "compound":
          prefix = prv_tok_text + " "+ tok.text
      
      # check: token is a modifier or not
      if tok.dep_.endswith("mod") == True:
        modifier = tok.text
        # if the previous word was also a 'compound' then add the current word to it
        if prv_tok_dep == "compound":
          modifier = prv_tok_text + " "+ tok.text
      
      ## chunk 3
      if tok.dep_.find("subj") == True:
        ent1 = modifier +" "+ prefix + " "+ tok.text
        prefix = ""
        modifier = ""
        prv_tok_dep = ""
        prv_tok_text = ""      

      ## chunk 4
      if tok.dep_.find("obj") == True:
        ent2 = modifier +" "+ prefix +" "+ tok.text
        
      ## chunk 5  
      # update variables
      prv_tok_dep = tok.dep_
      prv_tok_text = tok.text
  #############################################################
  entity_type = 'None'
  for element in ent2.split(" "):
    if element in entity_names:
      loc = entity_names.index(element)
      ent_label = entity_labels[loc]
      entity_type = 'None'
      if ent_label == 'NORP' or ent_label == 'ORG':
        entity_type = 'Personality'
      elif ent_label == 'GPE':
        entity_type = 'Location'
      else:
        entity_type = 'None'
  return [ent1.strip(), ent2.strip()], entity_type
```

```python
# Extracting the entity pairs of all sentences
entity_pairs = []
entity_ners = []

for i in tqdm(candidate_sentences["sentence"]):
  entities_list, entity_type = get_entities(i)
  entity_ners.append(entity_type)
  entity_pairs.append(entities_list)
```

```python
set(entity_ners)
location_counts = entity_ners.count('Location')
personality_counts = entity_ners.count('Personality')
print('Location counts', location_counts)
print('Personality counts', personality_counts)
```

```python
# List of target locations and personalities
location_targets = []
personality_targets = []
for i, entity_type in enumerate(entity_ners):
  if entity_type == 'Location':
    location_targets.append(entity_pairs[i][1])
  elif entity_type == 'Personality':
    personality_targets.append(entity_pairs[i][1])
print(location_targets)
print(personality_targets)
```

```python
# Function for obtaining the relations
def get_relation(sent):

  doc = nlp(sent)

  # Matcher class object 
  matcher = Matcher(nlp.vocab)

  #define the pattern 
  pattern = [{'DEP':'ROOT'}, 
            {'DEP':'prep','OP':"?"},
            {'DEP':'agent','OP':"?"},  
            {'POS':'ADJ','OP':"?"}] 

  matcher.add("matching_1", None, pattern) 

  matches = matcher(doc)
  k = len(matches) - 1

  span = doc[matches[k][1]:matches[k][2]] 

  return(span.text)
```

```python
# Extracting all relations
relations = [get_relation(i) for i in tqdm(candidate_sentences['sentence'])]
```

```python
# Creating the Knowledge Graphs
# Step:1 - Initializing the graph nodes and edjes 
# extract subject
source = [i[0] for i in entity_pairs]

# extract object
target = [i[1] for i in entity_pairs]

kg_df = pd.DataFrame({'source':source, 'target':target, 'edge':relations})
```

```python
# Creating the knowledge graph (directed graph)
# create a directed-graph from a dataframe
G=nx.from_pandas_edgelist(kg_df, "source", "target", 
                          edge_attr=True, create_using=nx.MultiDiGraph())
```

```python
# Visualizing the graph
plt.figure(figsize=(12,12))

pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_color='skyblue', edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```

```python
# Visualizaing only "written by" realtions
G1 = nx.from_pandas_edgelist(kg_df[kg_df['edge']=="written by"], "source", "target", 
                          edge_attr=True, create_using=nx.MultiDiGraph())

plt.figure(figsize=(12,12))
pos = nx.spring_layout(G1, k = 0.5)
nx.draw(G1, with_labels=True, node_color='skyblue', node_size=1500, edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```

# Centrality Measures

## PageRank and Betwness Centralities

```python
# Implementing centrality measures
from networkx.algorithms.link_analysis.pagerank_alg import pagerank
from networkx.algorithms.centrality import betweenness_centrality
```

```python
# Solution1 - Considering all targets while ranking
# PageRank
pg_cen = pagerank(G)
# Create a sub-dict with only "target" values
pg_cen_target_dict = {}
for key, value in pg_cen.items():
  if key in target:
    pg_cen_target_dict[key] = value
sort_pg_cen = sorted(pg_cen_target_dict, key=pg_cen_target_dict.get, reverse=True) # Sorting in the desceding order
print('Ranking Nodes based on PageRank Centrality')
print('Node\tValue')
for node in sort_pg_cen:
  print(node, pg_cen[node], sep='\t')
```

```python
# Solution1 - Considering all targets while ranking
# Betweeness Centrality
G_new = nx.DiGraph(G)
bw_cen = betweenness_centrality(G_new)
bet_cen_target_dict = {}
for key, value in bw_cen.items():
  if key in target:
    bet_cen_target_dict[key] = value
sort_bw_cen = sorted(bet_cen_target_dict, key=bet_cen_target_dict.get, reverse=True) # Sorting in the desceding order
print('Ranking Nodes based on Betweeness Centrality')
print('Node\tValue')
for node in sort_bw_cen:
  print(node, bw_cen[node], sep='\t')
```

```python
# Solution 2 - Considering Named entities (Location)
# PageRank
pg_cen = pagerank(G)
# Create a sub-dict with only "location target" values
pg_cen_loc_dict = {}
for key, value in pg_cen.items():
  if key in location_targets:
    pg_cen_loc_dict[key] = value
sort_pg_cen = sorted(pg_cen_loc_dict, key=pg_cen_loc_dict.get, reverse=True) # Sorting in the desceding order
print('Ranking Location Nodes based on PageRank Centrality')
print('Node\tValue')
for node in sort_pg_cen:
  print(node, pg_cen[node], sep='\t')
```

```python
# Solution 2 - Considering Named entities (Personalities)
# Create a sub-dict with only "location target" values
pg_cen_per_dict = {}
for key, value in pg_cen.items():
  if key in personality_targets:
    pg_cen_per_dict[key] = value
sort_pg_cen = sorted(pg_cen_per_dict, key=pg_cen_per_dict.get, reverse=True) # Sorting in the desceding order
print('Ranking Personality Nodes based on PageRank Centrality')
print('Node\tValue')
for node in sort_pg_cen:
  print(node, pg_cen[node], sep='\t')
```

```python
# Solution 2 - Considering Named entities (Location)
# Betwness centrality based on location
G_new = nx.DiGraph(G)
bw_cen = betweenness_centrality(G_new)
bet_cen_loc_dict = {}
for key, value in bw_cen.items():
  if key in location_targets:
    bet_cen_loc_dict[key] = value
sort_bw_cen = sorted(bet_cen_loc_dict, key=bet_cen_loc_dict.get, reverse=True) # Sorting in the desceding order
print('Ranking Location Nodes based on Betweeness Centrality')
print('Node\tValue')
for node in sort_bw_cen:
  print(node, bw_cen[node], sep='\t')
```

```python
# Solution 2 - Considering Named entities (Personality)
# Betwness centrality based on personality
G_new = nx.DiGraph(G)
bw_cen = betweenness_centrality(G_new)
bet_cen_per_dict = {}
for key, value in bw_cen.items():
  if key in personality_targets:
    bet_cen_per_dict[key] = value
sort_bw_cen = sorted(bet_cen_per_dict, key=bet_cen_per_dict.get, reverse=True) # Sorting in the desceding order
print('Ranking Personality Nodes based on Betweeness Centrality')
print('Node\tValue')
for node in sort_bw_cen:
  print(node, bw_cen[node], sep='\t')
```
